```python
import pandas as pd
```

```python
df = pd.read_csv("olympic_results.csv")
df
```

```python
# Filtrar apenas linhas com medalhas (remover linhas onde medal_type é vazio)
df_medalhas = df[df['medal_type'].notna()]

# Extrair o ano das Olimpíadas do slug_game (assumindo formato 'cidade-ano')
df_medalhas['year'] = df_medalhas['slug_game'].str.extract(r'-(\d{4})')[0]

# Agrupar por país, tipo de medalha e ano, e contar as ocorrências
medalhas_por_pais = df_medalhas.groupby(['country_name', 'country_code', 'slug_game', 'year', 'medal_type']).size().reset_index(name='count_medals')

# Ordenar os resultados
medalhas_por_pais = medalhas_por_pais.sort_values(['country_name', 'country_code', 'year', 'medal_type', 'slug_game'])
```

```python
# Salvar em um novo arquivo CSV
medalhas_por_pais.to_csv('../static/medals_by_country.csv', index=False)
```
